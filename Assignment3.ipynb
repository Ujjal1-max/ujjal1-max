```python
#Write a Python program that asks the user to input their age and prints whether they are eligible to vote (18 years or older) or not.
age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
#Write a Python program that checks if a user’s input number is positive, negative, or zero.
num = float(input("Enter a number: "))

if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive Number")
else:
    print("Negative Number")
```

```python
#write a program to determine if a number entered by the user is even or odd. The program should print 'Even Number' or 'Odd Number' based on the input.
num = int(input("Enter a number: "))

result = "Even Number" if num % 2 == 0 else "Odd Number"

print(result)
```

```python
#Write a Python program that takes the user’s age and prints which age group they belong to:
age = int(input("Enter your age: "))

if 0 <= age <= 12:
    print("Child")
elif 13 <= age <= 19:
    print("Teenager")
elif 20 <= age <= 64:
    print("Adult")
elif age >= 65:
    print("Senior")
else:
    print("Invalid age entered.")
```

```python
#Write a Python function sum_of_natural_numbers(n) that returns the sum of all natural numbers up to n. Use a loop inside the function.
def sum_of_natural_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

# Example usage:
num = int(input("Enter a number: "))
print("Sum of natural numbers up to", num, "is:", sum_of_natural_numbers(num))
```

```python
#Write a Python function factorial(n) that calculates and returns the factorial of n. Use a loop to calculate the factorial.
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

# Example usage:
num = int(input("Enter a number: "))
print("Factorial of", num, "is:", factorial(num))
```

```python
#Write a Python function multiplication_table(n) that prints the multiplication table of n up to 10.
def multiplication_table(n):
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")

# Example usage:
num = int(input("Enter a number: "))
multiplication_table(num)
```
